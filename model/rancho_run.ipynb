```python
# fs = glob.glob(os.path.join(parent,'result*'))
import numpy as np
import glob as glob
parent = '/content/runs'
import os
fs = glob.glob(os.path.join(parent,'test_run*'))

daymonth = [item.split('/')[-1].split('_')[2] for item in fs]
days = [item[0:2] for item in daymonth]
months = [item[2:4] for item in daymonth]
times = [item.split('/')[-1].split('_')[3][0:4] for item in fs]
dates = [pd.to_datetime( '2025-' + months[i] + '-' + days[i] + ' ' + times[i] ) for i in range (len(times))]
idx = np.argmax(dates)
cf = fs[idx]

epoch = 50
run_dir = Path(cf)
eval_run(run_dir=run_dir, period = 'test', epoch=epoch)
# eval_run(run_dir=run_dir, period = 'validation', epoch=epoch)
```

```python
model_str = "model_epoch0%d"%epoch
with open(run_dir / "test" / model_str / "test_results.p", "rb") as fp:
    results = pickle.load(fp)

# with open(run_dir / "validation" / "model_epoch050" / "validation_results.p", "rb") as fp:
#     validation = pickle.load(fp)
```

```python
import seaborn as sns
import pandas as pd
import xarray as xr
import numpy as np

var = 'ET'
obs = '%s_obs'%var
sim = '%s_sim'%var
timestep = '1D'

start = '2002-10-01'
stop = '2024-10-01'

for site in list(results):
  # extract observations and simulations

  path = '/content/drive/MyDrive/research/PML_LSTM/data/test_generic/time_series/%s.nc4'%site
  temp = xr.open_dataset(path).to_dataframe()
  # temp.loc['2021':,'ET'] = np.nan

  simvar = results[site][timestep]['xr'][sim].to_dataframe()
  # simvar[sim] = (10**simvar[sim]-0.1)**2
  simvar.index = pd.to_datetime([item[0] for item in simvar.index])

  fig, ax = plt.subplots(figsize=(8,3))
  temp['ET'].plot(ax=ax,label='Observed',ls='-')
  simvar['%s_sim'%var].plot(ax=ax,label='Simulated',ls='--')

  ax.legend()

  sns.despine()
  ax.legend(loc='upper right',fontsize=10)
  ax.set_ylabel('ET (mm/day)')
  ax.set_title(f"NSE {results[site][timestep]['NSE']:.3f}")
  # ax.set_title(f"RMSE {results[site][timestep]['RMSE']:.3f}")

  ax.set_xlim(start,stop)
  ax.set_ylim(0,4.5)
  sns.despine()
  fig.tight_layout()


ppt = temp['ppt'].loc[start:stop]

PML_predict = simvar['%s_sim'%var].loc[start:stop]
PML_predict = (PML_predict
    .reindex(ppt.index.union(PML_predict.index))
    .interpolate(method='time')
    .reindex(ppt.index)).interpolate()

defdf = pd.DataFrame({'P':ppt.values,'ET':PML_predict.values},index=PML_predict.index).dropna()
deficit = [0]
counter = 0
# Wang-Erlandsson deficit calculation
for idx,row in defdf.iterrows():
  A = row.ET - row.P
  next = np.max([0, deficit[counter] + A])
  deficit.append(next)
  counter = counter + 1
defdf['D'] = deficit[1:]
```

```python
defdf.to_csv('/content/defdf.csv')
```

```python

ssoil = 100
soil = [0]
rock = [0]
deficit = [0]

counter = 0
# Wang-Erlandsson deficit calculation
for idx,row in defdf.iterrows():
  A = row.ET - row.P

  if A>0:
    if soil[counter] >= ssoil:
      newrock = np.max([0, rock[counter] + A])
      soil.append(soil[counter])
      rock.append(newrock)
    else:
      rock.append(rock[counter])
      soil.append(np.max([0, soil[counter] + A]))
  else:
    soiltemp = soil[counter] + A
    if soiltemp<0:
      soiltemp = 0
      leftover = soil[counter] + A
    else:
      leftover = 0
    rocktemp = np.max([0,rock[counter] + leftover])
    soil.append(soiltemp)
    rock.append(rocktemp)

  next = np.max([0, deficit[counter] + A])
  deficit.append(next)
  counter = counter + 1


defdf['dsoil'] = soil[1:]
defdf['drock'] = rock[1:]
defdf['D'] = deficit[1:]

defdf['S'] = defdf['D'].max() - defdf['D']
defdf['ssoil'] = ssoil - defdf['dsoil']
defdf['srock'] = defdf['S'] - defdf['ssoil']


f,ax = plt.subplots(figsize=(8,3))


ax.fill_between(defdf.index, 0, defdf['S'],color='dodgerblue',label='Soil storage (Max = %d mm)'%int(ssoil))
ax.fill_between(defdf.index, 0, defdf['srock'],color='burlywood',label='Rock storage')
ax.legend(fontsize=8)

defdf['S'].plot(ax=ax,c='k')
defdf['srock'].plot(ax=ax,c='darkgoldenrod')

sns.despine()
ax.set_xlim('2019','2024-10')
ax.set_ylabel('Storage (mm)')
ax.set_xlabel('')

f.tight_layout()
```
